```python
#1 How many distinct (different) actors' last names are there?
engine.execute("USE sakila")
query = """SELECT COUNT(distinct(last_name)) as AmountofLastNames FROM sakila.actor;"""
firstquestion = pd.read_sql_query(query, engine)
firstquestion
```

```python
# 2 Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
```

```python
query= ''' SELECT*, 
 CASE
 WHEN DayofWEEK (rental_date) IN (1,2,3,4,5) then 'workday'
 WHEN DayofWeek (rental_date) IN (6,7) then 'weekend'
 ELSE 'No status'
 END AS 'Typeday'
 FROM rental;'''
data_query= pd.read_sql_query(query, engine)
data_query
```

```python
# 3 Get all films with ARMAGEDDON in the title.
```

```python
query = """SELECT * FROM sakila.film WHERE INSTR(title, 'ARMAGEDDON') > 0;"""
data_armageddon = pd.read_sql_query(query, engine)
data_armageddon
```

```python
#4 Get 10 the longest films.
```

```python
query= """SELECT * FROM film
ORDER BY length DESC
LIMIT 10;"""
data_longest= pd.read_sql_query(query, engine)
data_longest
```

```python
#5 How many films include Behind the Scenes content?
```

```python
query= """SELECT * FROM film
WHERE special_features IN ('Behind the Scenes');"""
data_scenes= pd.read_sql_query(query, engine)
data_scenes
```

```python
#6 Which kind of movies (rating) have a mean duration of more than two hours?
```

```python
query= """ SELECT rating, AVG(length) FROM film
group by rating
HAVING avg(length) > 120;"""
data_mean2= pd.read_sql_query(query, engine)
data_mean2
```

```python
#7 Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.
```

```python
query = """ SELECT length,title,RANK() OVER(ORDER by length DESC) AS ranked FROM film
WHERE length is NOT NULL AND length > 0
ORDER BY length DESC; """
data_rank= pd.read_sql_query(query, engine)
data_rank
```
